```python
import requests
import json

url_base = "https://loteriascaixa-api.herokuapp.com/api/"
```

```python
response = requests.get(url_base)
response.raise_for_status()
response = response.json()

print("API de Loterias da Caixa Econômica Federal - Todos os Jogos")
print(response)
```

```python
response = requests.get(url_base+'quina/latest')
response.raise_for_status()
response = response.json()

print("API de Loterias da Caixa Econômica Federal - Todos os Jogos")
print(response)

print("\n\nFormato JSON Pretty")
print(json.dumps(response, indent=4, sort_keys=True, ensure_ascii=False))
```

```python
response = requests.get(url_base+'quina/6584')
response.raise_for_status()
response = response.json()

print("API de Loterias da Caixa Econômica Federal - Todos os Jogos")
print(response)

print("\n\nFormato JSON Pretty")
print(json.dumps(response, indent=4, sort_keys=True, ensure_ascii=False))
```
